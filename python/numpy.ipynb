```python
# vectorization & easier operation
l1 = list(range(1,11))
l2 = list(range(11,21))

# 각 원소에 2를 곱한 배열
a = np.array(l1)
print(2 * a)

# 2 * a + b
b = np.array(l2)
print(2 * a + b, type(a), type(b))

# boolean operation
print(a == 5, b > 15)
```

```python
# 2-dimensional
c = np.array([[1,2,3],[10,11,12]])
print(len(c), len(c[0])) # 2, 3
print(c.ndim, c.shape) # 2 (2, 3)


# 3-dimensional
d = np.array([[[1,2,3],[10,11,12]],[[1,2,3],[10,11,12]],[[1,2,3],[10,11,12]],[[1,2,3],[10,11,12]]])
print(len(d), len(d[0]), len(d[0,1])) # 4, 2, 3
print(d.ndim, d.shape) # 3 (4, 2, 3)
```

```python
# slicing
test = np.array([[0, 1, 2, 3], [4, 5, 6, 7]])
print(test[0,:])	# 첫번째 행 전체
print(test[:,1])	# 두번째 열 전체
print(test[1,1:])	# 두번째 행의 두번째 열부터 끝 열까지
print(test[:2,:2])	# 2열까지
```

```python
a = np.arange(0,12).reshape(6,-1)
print(a)

# 전치연산
## T (transpose)
print(a.T)

# reshape
print(a.reshape(3,-1)) # -1 필요한 수를 자동으로 넣어준다
print(a.reshape(2,3,-1))

# flatten, ravel
print(a.flatten(), a.ravel())

# not equal! 
print(np.arange(5)) # 5
print(np.arange(5).reshape(1,5)) # (1,5)
print(np.arange(5).reshape(5,1)) # (5,1)

# 차원증가 (5 → (5,1))
print(np.arange(5)[:,np.newaxis])
```

```python
# 배열연결

## hstack (horizontal) : (2x3) * (2x2) -> (2x5)
a1 = np.ones((2,3)); print(a1)
a2 = np.zeros((2,2)); print(a2)
print(np.hstack([a1,a2]))

## vstack (vertical) : (2x3) * (3x3) -> (5x3)
b1 = np.zeros((3,3)); print(b1)
print(np.vstack([a1,b1]))

# dstack (depth) : (4x5) * (4x5) -> (4x5x2)
c1 = np.zeros((4,5)); print(c1)
c2 = np.ones((4,5))
print(np.dstack([c1,c2]))

# stack (depth, with sepcific direciton)
print(np.stack([c1,c2]))          # shape : (2,4,5)
print(np.stack([c1,c2], axis=1))  # shape : (4,2,5)
print(np.stack([c1,c2], axis=2))  # shape : (4,5,2)

# r_ (indexer) : vstack 동일?!
print(np.r_[a1,b1])

# c_

#tile

```

```python
# broadcasting : scala, 또는 저차원 행렬을 연산에 맞게 반복 확장해줌

# scala 
a = np.arange(100).reshape(10,-1)
print(a*10-5)

# (2,3) + (2,1)
print(np.arange(6).reshape(2,-1) + np.array([[10],[20]]))

# (2,3) + (1,3)
print(np.arange(6).reshape(2,-1) + np.array([10,20,30]))
```

```python
# dimension reduction

# 전체 통계
a = (10*np.arange(10)+1).reshape(2,-1)
print(a)
print(np.sum(a), a.sum())
print(np.min(a), a.min(), np.argmin(a), a.argmin()) # 최솟값, 최솟값 위치
print(np.max(a), a.max(), np.argmax(a), a.argmax()) # 최댓값, 최댓값 위치
print(np.mean(a), a.mean())
print(np.median(a))

#  축 중심 통계
print(np.sum(a, axis=0), np.sum(a, axis=1))
print(np.min(a, axis=0), np.min(a, axis=1))
print(np.max(a, axis=0), np.max(a, axis=1))
print(np.mean(a, axis=0), np.mean(a, axis=1))
print(np.median(a, axis=0), np.median(a, axis=1))

# boolean
b = np.zeros((100,100), dtype=np.int_)
print(np.any(b == 1), (b==1).any())
print(np.all(b == 0), (b==0).all())
```

```python
# sorting
a = np.array([[4,  3,  5,  7],
              [1, 12, 11,  9],
              [2, 15,  1, 14]])

print(np.sort(a)) # axis=1 동일, 가로방향 정렬
print(np.sort(a, axis=0)) # 세로방향 정렬
print(np.sort(a, axis=1))
print(np.argsort(a)) # 정렬순서
```
